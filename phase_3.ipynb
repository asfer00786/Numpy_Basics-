```python
from PIL import Image
import numpy as np


try:

        
    # Step 1: Open PNG
    image = Image.open("numpy.png")

    # Convert into Array 
    arr_image = np.array(image)

    # Step 3: Save as .npy
    np.save('image_numpy_logo.npy', arr_image)

    loaded_arr = np.load("image_numpy_logo.npy")
    print("Shape:", loaded_arr.shape)
    print("Data type:", loaded_arr.dtype)

    
    # Display
    logo = np.load('numpy-logo.npy')
    plt.figure(figsize=(10, 5))
    plt.subplot(121) #subplot(rows, cols, position)(1 row, 2 columns, 1 position)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    #Dark Logo
    #Assumption: logo has pixel values in the range 0–1 (float format). 0 = black 1 = white
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark Logo")
    plt.show()

except FileNotFoundError:
    print("numpy logo file not found")
```
