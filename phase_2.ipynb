# Practice With Resturants Data yearly

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Azeria(Vokeciu)
    [2, 120000, 140000, 160000, 190000],  # Moss pylimo
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Mcdolads 
])

print("=====  Sales Analysis  ======")
print("\n Sales Of Data Shape: ", sales_data.shape)
print("\n 1 to 3rd Data: ", sales_data[:4])
```

```python
# Finding sales per year why because resturanut onwer is same for all rest
per_yearly_sales = np.sum(sales_data[:,1:], axis=0)
print("Data Per Year of all Rest:", per_yearly_sales)

# Minimum sales per restaurant
min_sales  = np.min(sales_data[:,1:], axis=1)
print("Minimum sales per restaurant:", min_sales)

# Maximum sales per year
max_sales  = np.max(sales_data[:,1:], axis=0)
print("Maximum sales per year:", max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)

```

```python
# Dot Product and also find angles sin, cosine and tangent 

A = np.array([1, 0])
B = np.array([0, 1])

dot_product = np.dot(A, B)
magnitude = np.linalg.norm(A) * np.linalg.norm(B)

angle_rad = np.arccos(dot_product / magnitude)
angle_deg = np.degrees(angle_rad)

print(angle_deg)  # 90.0 degrees


# Forward: angle → sine value
print(np.sin(np.radians(30)))   # 0.5

# Reverse: sine value → angle
print(np.degrees(np.arcsin(0.5)))  # 30.0 degrees

```
